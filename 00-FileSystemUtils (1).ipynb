####File System Utility (dbutils.fs)

```
# help for ls, cp, mv, rm, head, mkdirs
dbutils.fs.help("cp")
```

```
# show tables in folder
# #dbutils.fs.ls("/FileStore/tables/[<folder-name>]/[<file-name>]")
dbutils.fs.ls("/FileStore/tables/")
```

```
# copy file
# #dbutils.fs.cp("/FileStore/tables/[folder-name]/<source-file-name>","/FileStore/tables/[folder-name]/<target-file-name>")
dbutils.fs.cp("/FileStore/tables/catalog.csv","/FileStore/tables/new-catalog.csv")
```

```
# move file / ren file
# #dbutils.fs.mv("/FileStore/tables/[folder-name]/<source-file-name>","/FileStore/tables/[folder-name]/<target-file-name>")
dbutils.fs.mv("/FileStore/tables/new-catalog.csv","/FileStore/tables/old-catalog.csv")
```

```
# delete table
# dbutils.fs.rm("/FileStore/tables/[<folder-name>]/<file-name>")
dbutils.fs.rm("/FileStore/tables/old-catalog.csv")
```

```
# head
# dbutils.fs.head("/FileStore/tables/[<folder-name>]/<file-name>,[bytes]")
dbutils.fs.head("/FileStore/tables/catalog.json")
```

```
# mkdir
# dbutils.fs.mkdirs("/FileStore/tables/<folder-name>")
dbutils.fs.mkdirs("/FileStore/tables/junk")
```

```
# rmdir
# dbutils.fs.rm("/FileStore/tables/<folder-name>")
dbutils.fs.rm("/FileStore/tables/junk")
```
